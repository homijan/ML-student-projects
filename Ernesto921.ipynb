```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/fenago/datasets/main/Bias_correction_ucl.csv")
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
year_dummies = pd.get_dummies(df['Date'].dt.year,prefix='year')
year_dummies
```

```python
month_dummies = pd.get_dummies(df['Date'].dt.month,prefix='month')
month_dummies
```

```python
df = pd.concat([df, month_dummies, year_dummies], axis=1)
```

```python
df
```

```python
df.drop('Date', axis=1, inplace=True)
```
